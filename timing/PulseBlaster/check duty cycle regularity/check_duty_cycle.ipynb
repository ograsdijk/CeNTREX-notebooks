```python
import numpy as np
from nptdms import TdmsFile
import matplotlib.pyplot as plt
```

```python
tdms_file = TdmsFile("Instrument Oscilloscope Capture 2020-10-03 17-34-36.tdms")
tdms_file.groups()
```

```python
tdms_file.group_channels("Oscilloscope - Waveform Data")
```

```python
ch2 = tdms_file.channel_data("Oscilloscope - Waveform Data", 'Channel 2 (PXI1Slot2/2)')
ch3 = tdms_file.channel_data("Oscilloscope - Waveform Data", 'Channel 3 (PXI1Slot2/3)')
```

```python
sl = np.s_[0:30000]
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(ch2[sl], '.-')
ax.plot(ch3[sl], '.-')
```

```python
sl = np.s_[0:30000]
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(ch2[sl], '.-')
ax.plot(ch3[sl], '.-')
```

```python
indices_change = np.where(np.abs(np.diff(ch2)) > 1)[0]+1
len_low_high = np.diff(indices_change)

print(len_low_high[:25])

print(f'{1/(1/250e6 * 1250*2):.2e} Hz')
```

```python
sl = np.s_[0:100]
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(ch2[sl], '.-')
ax.plot(ch3[sl], '.-')

sl = np.s_[550:1000]
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(ch2[sl], '.-')
ax.plot(ch3[sl], '.-')
```

```python
high = np.sum(ch2 > 2.45)  
low  = np.sum(ch2 < 0.025)

print(f'{high/(high+low)*100:.2f}% high, {low/(high+low)*100:.2f}% low')
```
