```python
import numpy as np
from scipy.constants import c
from utils.plot_utils import *
import matplotlib.pyplot as plt
```

```python
α = np.deg2rad(1) # misalignment laser with molecular beam
Γ = 1/100e-9
f0 = c/271e-9
vz = 184
σvz = 16.8
```

```python
f0/1e12
```

```python
def resonance_shift(α, vz, f0):
    return np.sin(α)*(vz/c)*f0
```

```python
alphas = np.deg2rad(np.linspace(0,1,6))
fig, axes = plt.subplots(len(alphas), figsize = (8,15), sharex = True)

for idx, α in enumerate(alphas):    
    s = resonance_shift(α, vz, f0)
    axes[idx].axvline(-s/1e6, lw = 5)
    axes[idx].axvline(s/1e6, lw = 5)
    
    axes[idx].text(0.03, 0.95, f"α = {np.rad2deg(α):.1f}", transform = axes[idx].transAxes,
        ha = 'left', va = 'top', fontsize = 14);
    
    axes[idx].set_xlabel("Δ UV [MHz]")
    
    set_fontsize(axes[idx], 14)
```

```python
alphas = np.deg2rad(np.linspace(0,1,6))
fig, axes = plt.subplots(len(alphas), figsize = (8,15), sharex = True)

for idx, α in enumerate(alphas):    
    s = resonance_shift(α, vz, f0)
    s /= 4
    axes[idx].axvline(-s/1e6, lw = 5)
    axes[idx].axvline(s/1e6, lw = 5)
    
    axes[idx].text(0.03, 0.95, f"α = {np.rad2deg(α):.1f}", transform = axes[idx].transAxes,
        ha = 'left', va = 'top', fontsize = 14);
    
    axes[idx].set_xlabel("Δ IR [MHz]")
    
    set_fontsize(axes[idx], 14)
```

```python
alphas = np.deg2rad(np.linspace(0,1,6))
fig, axes = plt.subplots(len(alphas), figsize = (8,15), sharex = True)

for idx, α in enumerate(alphas):
    
    γ = Γ/(2*np.pi)
    
    s = resonance_shift(α, vz, f0)
    axes[idx].axvline(-s/γ, lw = 5)
    axes[idx].axvline(s/γ, lw = 5)
    
    axes[idx].text(0.03, 0.95, f"α = {np.rad2deg(α):.1f}", transform = axes[idx].transAxes,
        ha = 'left', va = 'top', fontsize = 14);
    
    axes[idx].set_xlabel("Δ [Γ]")
    
    set_fontsize(axes[idx], 14)
```
