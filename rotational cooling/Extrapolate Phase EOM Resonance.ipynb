```python
import numpy as np
import matplotlib.pyplot as plt
from utils.plot_utils import set_fontsize
```

```python
powers = [4, 6, 8, 10, 12, 14, 16] # dBm
resonance = [1.615, 1.614, 1.613, 1.611, 1.609, 1.607, 1.604] # MHz
```

```python
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(powers, resonance, '.', ms = 15)

ax.set_xlabel('power [dBm]')
ax.set_ylabel('resonance [MHz]')

set_fontsize(ax, 15)
```

```python
xfit = np.linspace(min(powers), 24, 101)
fit = np.poly1d(np.polyfit(powers, resonance,2))
```

```python
fig, ax = plt.subplots(figsize = (8,5))

ax.plot(xfit, fit(xfit))
ax.plot(powers, resonance, '.', ms = 15)
ax.plot(24, fit(24), 'C3.', ms = 15)

ax.text(0.95, 0.95, f'{fit(24):.3f} MHz @24 dBm', transform = ax.transAxes,
        va = 'top', ha = 'right', fontsize = 15)

ax.set_xlabel('power [dBm]')
ax.set_ylabel('resonance [MHz]')

set_fontsize(ax, 15)
```
