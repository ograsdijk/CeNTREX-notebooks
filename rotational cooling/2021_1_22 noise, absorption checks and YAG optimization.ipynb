## Some statistics on sets

```python
fname = "2021_1_22.hdf"

dset_names = [
                "1611353545 noise check with laser",
                "1611354872 noise check no laser lock",
                "1611355803 noise check laser piezo no lowpass",
                "1611355926 noise check laser piezo no lowpass not connected to adc",
                "1611355992 noise check laser wl mod off",
                "1611366142 noise check laser wl mod on 230 Hz RC filter"
            ]

names = [
    "1 kHz lowpass",
    "no laser lock",
    "no lowpass",
    "no NI DAQ",
    "WL mod off",
    "230 Hz lowpass"
]
```

```python
stds = []
ptps = []

for dset_name in dset_names:
    pmt_channel = 1
    pd_channel = 3

    pxie, pxie_time, pxie_attrs, data_devices = load_measurement_data_devices_attrs(fname, dset_name, devices)

    print(f'{len(pxie)} samples')

    pmt_data = np.array([pxie[idx][:,pmt_channel].astype(float) for idx in range(1,len(pxie)+1)])
    pd_data = np.array([pxie[idx][:,pd_channel].astype(float) for idx in range(1,len(pxie)+1)])
    
    stds.append(np.std(pd_data, axis = 1))
    ptps.append(np.ptp(pd_data, axis = 1))
```

```python
fig, axes = plt.subplots(nrows = 6, figsize = (8,10), sharex = True)
bins = np.arange(0,140,5)

for idx, (name, std) in enumerate(zip(names, stds)):
    axes[idx].hist(std, label = name, density = True, bins = bins, color = f'C{idx}')
    axes[idx].text(0.97, 0.9, name, transform = axes[idx].transAxes, fontsize = 14,
                  va = 'top', ha = 'right')
    set_fontsize(axes[idx], 14)

axes[-1].set_xlabel(' [adc]')
```

```python
for name, std in zip(names, stds):
    print(f"{name:<15} : {np.mean(std)}")
```

Here the negative values for the translation stage correspond to when the translation stage was moved 1" back, e.g. 0 before corresponds to 10 after. (9 = -1, 8 = -2, etc.)

```python
fname = "2021_1_21 2.hdf"
dset_name = "1611347417 YAG focal length optimization"
pmt_channel = 1
pd_channel = 3

pxie, pxie_time, pxie_attrs, data_devices = load_measurement_data_devices_attrs(fname, dset_name, devices)

print(f'{len(pxie)} samples')

dt = 1/1e5
n = pxie[1].shape[0]

t = np.arange(0,n,1)/100

pmt_data = np.array([pxie[idx][:,pmt_channel].astype(float) for idx in range(1,len(pxie)+1)])
pd_data = np.array([pxie[idx][:,pd_channel].astype(float) for idx in range(1,len(pxie)+1)])
```

```python
dset_len = pmt_data.shape[-1]
fl = -pmt_data.copy()
fl -= fl[:,:200].mean(axis = 1)[:,np.newaxis]
fl = fl.reshape(-1,50,10,dset_len)
```

```python
plt.plot(fl.mean(axis = 2)[1].T);
```

```python
focal_lengths = [
    4, 4.5, 5.5, 3, 2, 1, 0, 0, -1, -1, -2, -3, -4, -5, -6, -7, 0, -1
]
```

```python
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(focal_lengths, np.trapz(fl.mean(axis = 2).mean(axis = 1), axis = 1), '.', ms = 10)

ax.set_xlabel('translation stage lens')
ax.set_ylabel('fluorescence integral')

set_fontsize(ax, 14)
```

```python
integrals = []
for integral in np.trapz(fl.mean(axis = 2), axis = 2):
    tmp = []
    for val in integral:
        if val > 2e3:
            tmp.append(val)
    integrals.append(np.mean(tmp))
```

```python
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(focal_lengths, integrals, '.', ms = 10)

ax.set_xlabel('translation stage lens')
ax.set_ylabel('fluorescence integral')

set_fontsize(ax, 14)
```
