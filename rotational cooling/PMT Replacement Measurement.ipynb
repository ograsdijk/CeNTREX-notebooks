# Checking PMT Gain behaviour

measurements taking at different PMT voltages, measurements organized as shown in the table below

| | Laser 1 off | Laser 1 on | Laser 1&2 on | Laser 2 on |
|-----|-----|-----|-----|-----|
|Full lights / shielding | | | | 
| No lights / shielding | | | |
| Low lights / shielding | | | |

```python
pmt_measurements = {-800: np.array([[-13.55, -23.7, -27.7, -17.6],
                                    [-0.2, -9.7, -13.5, -4.1],
                                    [-0.3, -9.7, -13.7, -4.2]]),
                    -900: np.array([[-29.683, -54.64, -65.02, -38.5],
                                    [-0.2, -22.2, -31.23, -8.6],
                                    [-0.5, -22.3, -31.7, -8.9]]),
                    -1000: np.array([[-60.1, -117, -141.7, -79.8],
                                     [-0.32, -45.9, -66, -17.6],
                                     [-0.8, -46.5, -66, -18]]),
                    -1100: np.array([[-116.5, -248, -297.8, -164],
                                     [-0.45, -92.6, -131.4, -33.7],
                                     [-1.3, -93.5, -132.7, -34.4]])
                   }
desc_y = ['full lights', 'no lights', 'low lights']

desc_x = ['no lasers', 'laser 1', 'laser 1&2', 'laser 2']
colors = {-800: {'rot on': 'C0', 'rot off': 'C1'}, -900: {'rot on': 'C2', 'rot off': 'C3'}, -1000: {'rot on': 'C4', 'rot off': 'C5'}, -1100: {'rot on': 'C6', 'rot off': 'C7'}}
```

```python
fig, ax = plt.subplots(figsize = (10,6))
for voltage, data in pmt_measurements.items():
    for idx, (desc, dat) in enumerate(zip(desc_y, data)):
        ax.plot(idx, dat[2]-dat[1], '.', ms = 15, color = colors[voltage]['rot on'], label = f'{voltage} V: rot on')
        ax.plot(idx, dat[3]-dat[0], '.', ms = 15, color = colors[voltage]['rot off'], label = f'{voltage} V: rot off')
        
ax.set_xticks(range(len(desc_y)))
ax.set_xticklabels(desc_y)

ax.set_ylabel('PMT signal [mV]')
ax.set_title('PMT signal from laser 1 under different ambient lightning conditions')


box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
handles, labels = fig.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), loc='center left', bbox_to_anchor=(1, 0.5), fontsize = 15)

set_fontsize(ax, 15)
```

```python
colors = {-800: 'C0', -900: 'C1', -1000: 'C2', -1100: 'C3'}
```

```python
fig, ax = plt.subplots(figsize = (10,6))
for voltage, data in pmt_measurements.items():
    for idx, (desc, dat) in enumerate(zip(desc_y, data)):
        ax.plot(idx,  (dat[2]-dat[1]) / (dat[3]-dat[0]), '.', ms = 15, color = colors[voltage], label = f'{voltage} V')
        
ax.set_xticks(range(len(desc_y)))
ax.set_xticklabels(desc_y)

ax.set_ylabel('signal ratio')
ax.set_title('Ratio signal laser 1 with rotational cooling on or off')

handles, labels = fig.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), fontsize = 15)

set_fontsize(ax, 15)
```

```python
fig, ax = plt.subplots(figsize = (10,6))
for voltage, data in pmt_measurements.items():
    for idx, (desc, dat) in enumerate(zip(desc_y, data)):
        ax.errorbar(idx,  (dat[2]-dat[1]) / (dat[3]-dat[0]), yerr = 0.05, fmt = '.', ms = 15, color = colors[voltage], label = f'{voltage} V')
        
ax.set_xticks(range(len(desc_y)))
ax.set_xticklabels(desc_y)

ax.set_ylabel('signal ratio')
ax.set_title('Ratio signal laser 1 with rotational cooling on or off')

handles, labels = fig.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), fontsize = 15)

set_fontsize(ax, 15)
```
