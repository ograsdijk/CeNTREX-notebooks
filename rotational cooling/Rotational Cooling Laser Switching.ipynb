# Switching rotational cooling light

Switching the rotational cooling light on/off with a shutter, shutter state is captured in CH2 of the PXIe-5171 data acquisition. 

| high| low | 
|-----|-----|
|open|closed|

PMT is set to 1 kV, 2x factor 5 amplification stages on SRS amplifier, with 10 kHz lowpass filter.

PXIe-5171 Channels

|ch0  |ch1    |ch2    |
|------|------|------|
|trigger  |PMT | polarization or light state|

## R0 F2 detection, P2 F1 rotational cooling

```python
fname = "rotation_cooling.hdf"
dset_name = "1582245525 switching_rot_cooling_light_P2_F1_detection_R0_F2"
title = "R0 F2 detection, P2 F1 rotational cooling"

switch_state, bin_state_data, ratio = bin_dataset_switching(fname, dset_name, pmt_channel = 1, switching_channel = 2, threshold = 500)
plot_binned_dataset_switching(bin_state_data, ratio, title = title, switch_labels = {0: 'no rotational cooling', 1:'rotational cooling'})
```

# Measurements 27-2-2020


## R0 F1 detection, P2 F2 rotational cooling
Switching rotational cooling light on/off with a shutter.  

Double passed the detection light with a single prism. Detection light modulated with modulator (QUBIG AM7-UV, S/N: KA1059).  

PMT is set to 1 kV, 2x factor 5 amplification stages on SRS amplifier, with 10 kHz lowpass filter.  

| desc | value | 
|-----|-----|
|rot cooling pol type | square wave, cont.|
|rot cooling pol switch | 1.69 Mhz|
|rot cooling light state | switched |
|rot cooling power | 50 mW |
|detection pol type | resonant, cont. |
|detection pol switch | 1.562 MHz | 
|detection power | 29 mW |

PXIe-5171 Channels  

|ch0  |ch1    |ch2    |
|------|------|------|
|trigger  |PMT | rot. cooling light state|

ch2 shutter state

| high| low | 
|-----|-----|
|open|closed|
