## R0 F1 detection, P2 F2 rotational cooling
Switching detection laser polarization.

Single passed the detection light with a single prism. Detection light modulated with modulator (QUBIG AM7-UV, S/N: KA1059), only 0.6$\pi$ modulation.

Multipassed the rotational cooling light with two prisms.

PMT is set to 1 kV, 2x factor 5 amplification stages on SRS amplifier, with 10 kHz lowpass filter.  

| desc | value | 
|-----|-----|
|rot cooling pol type | square wave, gated|
|rot cooling pol switch | 1.69 Mhz|
|rot cooling light state | on |
|rot cooling power | 40 mW |
|detection pol type | resonant, cont. |
|detection pol switch | 1.562 MHz | 
|detection power | 32 mW |

PXIe-5171 Channels  

|ch0  |ch1    |ch2    |
|------|------|------|
|trigger  |PMT | rot. cooling pol state|

ch2 pol state

| high| low | 
|-----|-----|
|switching on|switching off|
