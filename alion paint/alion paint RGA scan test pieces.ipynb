```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname = '10_empty_bellows2.txt'
pump_empty = np.loadtxt(fname, delimiter = ',', skiprows = 22, usecols = (0,1))

fig, ax = plt.subplots(figsize = (8,5))
ax.plot(pump_empty[:,0], pump_empty[:,1])
ax.set_yscale('log')
ax.set_xlabel('mass [amu]')
ax.set_ylabel('[torr]')
ax.set_title('RGA scan empty chamber')
ax.set_ylim(ymin = 1e-10);
```

```python
fname = 'alionpaint_pumping_3hours.txt'
pump_3hours = np.loadtxt(fname, delimiter = ',', skiprows = 22, usecols = (0,1))

fig, ax = plt.subplots(figsize = (8,5))
ax.plot(pump_3hours[:,0], pump_3hours[:,1])
ax.set_yscale('log')
ax.set_xlabel('mass [amu]')
ax.set_ylabel('[torr]')
ax.set_title('RGA scan alion paint pumping 3 hours')
ax.set_ylim(ymin = 1e-10);
```

```python
fname = 'alionpaint_pumping_overnight.txt'
pump_overnight = np.loadtxt(fname, delimiter = ',', skiprows = 22, usecols = (0,1))

fig, ax = plt.subplots(figsize = (8,5))
ax.plot(pump_overnight[:,0], pump_overnight[:,1])
ax.set_yscale('log')
ax.set_xlabel('mass [amu]')
ax.set_ylabel('[torr]')
ax.set_title('RGA scan alion paint pumping overnight')
ax.set_ylim(ymin = 1e-10);
```

```python
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(pump_empty[:,0], pump_empty[:,1], label = 'empty')
ax.plot(pump_3hours[:,0], pump_3hours[:,1], label = '3h')
ax.set_yscale('log')
ax.set_xlabel('mass [amu]')
ax.set_ylabel('[torr]')
ax.set_ylim(ymin = 1e-10)
ax.set_xlim(xmax = 65)
ax.set_title('RGA scan empty vs alion paint')
ax.legend();
```

```python
fig, ax = plt.subplots(figsize = (8,5))
ax.plot(pump_3hours[:,0], pump_3hours[:,1], label = '3h')
ax.plot(pump_overnight[:,0], pump_overnight[:,1], label = 'overnight')
ax.set_yscale('log')
ax.set_xlabel('mass [amu]')
ax.set_ylabel('[torr]')
ax.set_ylim(ymin = 1e-10)
ax.set_title('RGA scan alion paint')
ax.legend();
```
