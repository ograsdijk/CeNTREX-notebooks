```python
import json
import pickle
import numpy as np
import matplotlib.pyplot as plt
from utils.plot_colormesh import *
import matplotlib.animation as animation
from utils.grid_plot_trajectories import *

run_name = 'straight_trajectory_scan_homogeneous_field_B_field'

options_dir = 'options/'
results_dir = ''
plots_dir = ''

results_fname = 'straight_trajectory_scan_homogeneous_field_B_field-4df0ae36f4a2be5a723c5b2b9c1b088e.txt'
```

```python
with open(options_dir + f'{run_name}.json') as json_file:
    options = json.load(json_file)
    
with open(results_dir + results_fname) as results_file:
    results = np.loadtxt(results_file)
    
pickled_vars = {}
for key, fname in options['pickle_fnames'].items():
    fname = options_dir + fname.split('/')[-1]
    with open(fname, 'rb') as f:
        pickled_vars[key] = pickle.load(f)
```

```python
plot_options = {'levels':[], 'xlabel': 'Bz [G]', 'ylabel': 'x84 [V/cm]', 'scale':'linear'}
plot_pmesh(results[:,1],results[:,2],results[:,-1], **plot_options)
```

```python
plot_options = {'levels':[], 'xlabel': 'Bz [G]', 'ylabel': 'x84 [V/cm]', 'scale':'log'}
plot_pmesh(results[:,1],results[:,2],results[:,-1], **plot_options)
```
